### analyze NBA results provided by FiveThirtyEight in a 17MB CSV file

```python
# downloading dataset

import requests

download_url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/nba-elo/nbaallelo.csv"
target_csv_path = "nba_all_elo.csv"

response = requests.get(download_url)
response.raise_for_status()    # Check that the request was successful
with open(target_csv_path, "wb") as f:
    f.write(response.content)
print("Download ready.")
```

```python
# read the csv file as pandas dataframe

import pandas as pd
nba = pd.read_csv("nba_all_elo.csv")
```

### Missing values

#### method 1: drop the rows containing the missing values

```python
rows_without_missing_data = nba.dropna()
rows_without_missing_data.shape
```

```python
rows_without_missing_data
```

it will lose lots of observations as rowwise

#### Method 2 : drop the particular column

```python
data_without_missing_columns = nba.dropna(axis=1)
data_without_missing_columns.shape
```

```python
data_without_missing_columns
```

#### Method 3: fill the none values with some data

```python
data_with_default_notes = nba.copy()
data_with_default_notes["notes"].fillna(value="no notes at all",inplace=True)
data_with_default_notes["notes"].describe()
```
